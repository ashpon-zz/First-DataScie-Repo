```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
# Read Data 
cityFile = "C:/Rutgers/python/homework/PandasNumpy/Resources/city_data.csv"
rideFile = "C:/Rutgers/python/homework/PandasNumpy/Resources/ride_data.csv"
```

```python
# Create DataFrame from CityCSV
cityDf = pd.read_csv(cityFile)
cityDf.set_index('city')
#cityDf.head(5)
```

```python
# Create DataFrame from rideCSV
rideDf = pd.read_csv(rideFile)
rideDf.set_index('city')
#rideDf.get_value('Sarabury','date')
```

```python
#avgFare = rideDf.pivot_table(['fare'], index='city').reset_index()
avgTotFare = rideDf.pivot_table(index='city', aggfunc={'ride_id':'count', 'fare':'mean'}).reset_index()
avgTotFare.columns = ['city','AvgFare','TotRides']
#avgTotFare.head(5)
```

```python
cityRides = pd.merge(cityDf,avgTotFare,how='outer')
#cityRides.head(5)
```

```python
plt.figure(figsize=(16, 6))
plt.grid = True
plt.scatter(cityRides["TotRides"].loc[cityRides["type"] == 'Urban'],
            cityRides['AvgFare'].loc[cityRides["type"] == 'Urban'],
            cityRides['driver_count'].loc[cityRides["type"] == 'Urban']*4,
            marker = 'o',facecolors="gold", edgecolors="black", alpha=0.75,label="Urban", linewidth='0.7')
plt.scatter(cityRides["TotRides"].loc[cityRides["type"] == 'Rural'],
            cityRides['AvgFare'].loc[cityRides["type"] == 'Rural'],
            cityRides['driver_count'].loc[cityRides["type"] == 'Rural']*4,
            marker = 'o',c = 'lightskyblue', edgecolors="black", alpha=1,label="Rural", linewidth='0.7')
plt.scatter(cityRides["TotRides"].loc[cityRides["type"] == 'Suburban'],
            cityRides['AvgFare'].loc[cityRides["type"] == 'Suburban'],
            cityRides['driver_count'].loc[cityRides["type"] == 'Suburban']*4,
            marker = 'o',c ="lightcoral", edgecolors="black", alpha=0.90,label="Suburban", linewidth='0.7')
plt.xlabel("Total Rides")
plt.ylabel("Average Fare")
plt.legend()
plt.title('Total Rides Vs Average Fare by City Type')
plt.show()
```

```python
#Pie Charts
# % of Total Fares by City Type
#Calculate totals
totFares = rideDf["fare"].sum()
totRides = cityRides["TotRides"].sum()
totDrivers = cityDf["driver_count"].sum()
mergedDf = pd.merge(cityDf,(rideDf.pivot_table(index='city', aggfunc={'ride_id':'count', 'fare':'sum'}).reset_index()))
grpdTots = mergedDf.groupby(["type"]).sum()
percUrbanFare = grpdTots.get_value('Urban','fare')/totFares
percRuralFare = grpdTots.get_value('Rural','fare')/totFares
percSubFare   = grpdTots.get_value('Suburban','fare')/totFares
# The colors of each section of the pie chart
colors = ["gold",  "lightskyblue", "lightcoral"]
plt.pie([percRuralFare,percSubFare,percUrbanFare], labels=['Rural Fare','Suburban Fare', 'Urban Fare'], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=200)
plt.title(" % of Total Fare by City Type")
plt.show()
```

```python
# % of Total Drivers by City Type
percUrbanDrivers = grpdTots.get_value('Urban','driver_count')/ totDrivers
percRuralDrivers = grpdTots.get_value('Rural','driver_count')/totDrivers
percSubDrivers   = grpdTots.get_value('Suburban','driver_count')/totDrivers
# The colors of each section of the pie chart
colors = ["gold",  "lightskyblue", "lightcoral"]
plt.pie([percRuralDrivers,percSubDrivers,percUrbanDrivers], labels=['Rural Drivers','Suburban Drivers', 'Urban Drivers'], colors=colors,
         autopct="%1.1f%%", shadow=True, startangle=200)
plt.title(" % of Total Drivers by City Type")
plt.show()
```

```python
#% of Total Rides by City Type
mergedDf1 = pd.merge(cityDf,rideDf)
grpdCount = mergedDf1[["type","ride_id"]].groupby(["type"]).count()
percUrbanRides = grpdCount.get_value('Urban','ride_id')/totRides
percRuralRides = grpdCount.get_value('Rural','ride_id')/totRides
percSubRides   = grpdCount.get_value('Suburban','ride_id')/totRides
# The colors of each section of the pie chart
colors = ["gold",  "lightskyblue", "lightcoral"]
plt.pie([percRuralRides,percSubRides,percUrbanRides], labels=['Rural Rides','Suburban Rides', 'Urban Rides'], colors=colors,
         autopct="%1.1f%%", shadow=True, startangle=200)
plt.title(" % of Total Rides by City Type")
plt.show()
#grpdCount
```
